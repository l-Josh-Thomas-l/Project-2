```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('default')
```

```python
data ="../data/fertilizer-use-in-kg-per-hectare-of-arable-land.csv"
```

```python
data_df = pd.read_csv(data, encoding="ISO-8859-1")

data_df.head(5)
```

```python
data_df = data_df.rename(columns={'Entity':'Country/Continent'})
data_df=data_df.drop('Code',1);

data_df
```

```python
Aust_df = data_df.loc[data_df["Country"] == "Australia", :]

Aus_df.head(5)
```
